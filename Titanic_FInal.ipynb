apply 메서드가 더 쉬움

```python
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
df_train['Age_cat_2'] = df_train['Age'].apply(category_age) # lambda 비슷
```

```python
(df_train['Age_cat'] == df_train['Age_cat_2']).all() # all은 모두 True여야 하고, any는 1개라도 True면 True 반환
```

```python
from sklearn.ensemble import RandomForestClassifier # decision tree를 random으로 앙상블(average)시킨 거
from sklearn import metrics
from sklearn.model_selection import train_test_split
# 파이썬을 활용한 머신러닝 책 좋음
```

```python
X_train = df_train.drop('Survived', axis=1).values
target_label = df_train['Survived'].values
X_test = df_test.values
```

```python
X_tr, X_vld, y_tr, y_vld = train_test_split(X_train, target_label, test_size=0.3, random_state=2018) # y_tr을 줬으니 지도학습  
```

```python
model = RandomForestClassifier()
model.fit(X_tr, y_tr)
```

```python
prediction = model.predict(X_vld)
```

```python
print('총 {}명 중 {:.2f}% 정확도로 생존 맞춤'.format(y_vld.shape[0], 100*metrics.accuracy_score(prediction, y_vld)))
```
